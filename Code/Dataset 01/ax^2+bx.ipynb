```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler
from scipy import stats
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
#prepare data
data = pd.read_csv('_Datapreprocessing.csv', skiprows = 2, sep=",") #remove 2 rows
data.drop('1', axis = 1, inplace = True)
```

```python
x = data.iloc[:, 1:6]
y = data.iloc[:, 6]
```

```python
#for missing values
imputerr = KNNImputer(n_neighbors = 2)
x.iloc[:, 1:4] = imputerr.fit_transform(data.iloc[:, 2:5])
```

```python
#the whole data
pd.DataFrame(x)
pd.DataFrame(y)
not_nan_data = x.join(y)
not_nan_data
```

```python
x2 = pd.DataFrame(np.square(not_nan_data['Total population']))
x2.rename(columns = {'Total population': 'x^2'}, inplace = True)
x2
```

```python
x_train = x2.join(pd.DataFrame(not_nan_data.iloc[:, 2]))
x_train.rename(columns = {'Total population': 'x'}, inplace= True)
x_train
```

```python
y_train = pd.DataFrame(not_nan_data.iloc[:, 5])
y_train
```

```python
lm = LinearRegression()
lm.fit(x_train, y_train)
```
