```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
%matplotlib inline
```

```python
data = pd.read_csv('arrhythmia_csv.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
#missing values
impute = KNNImputer(n_neighbors = 2)
data = pd.DataFrame(impute.fit_transform(data), columns=data.columns)
data
```

```python
sns.countplot(x = 'class', data = data)
```

## stndard variables

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data.drop('class', axis= 1))
scaled_features
```

```python
normal_data = pd.DataFrame(scaled_features, columns=data.columns[:-1])
normal_data
```

```python
x_train, x_tets, y_train, y_test = train_test_split(normal_data, data['class'], test_size= 0.3 )
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 30, metric = 'euclidean')
knn.fit(x_train, y_train)
y_pred = knn.predict(x_tets)
y_pred_train = knn.predict(x_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
#train
print(confusion_matrix(y_train, y_pred_train))
print(classification_report(y_train, y_pred_train))
```

```python
print(classification_report(y_test, y_pred))
```
